```python
import requests
import bs4
```

```python
url = "https://books.toscrape.com/catalogue/page-2.html"
response = requests.get(url)
response.content
```

```python
soup = bs4.BeautifulSoup(response.content)
soup
```

```python
s1 = soup.findAll('article',{'class':'product_pod'})[2]
print(s1.img.attrs["src"])
print(s1.h3.a.attrs['title'].replace(',',' '))
s2 = soup.findAll('div',{"class":"product_price"})[2]
print(s2.p.text)
```

```python
#standalone function:

with open("bookstore.csv",'w', encoding='utf-8') as f:
    f.write("image_url,book_title,product_price\n")
    for i in range(0,50):
        url = "http://books.toscrape.com/catalogue/page-{}.html".format(i+1)
        response = requests.get(url)
        soup = bs4.BeautifulSoup(response.content)
        s1=soup.findAll('article',{'class':'product_pod'})
        s2=soup.findAll('div',{"class":"product_price"})
        for x in range(len(s1)):
            a=s1[x].img.attrs['src']
            b=s1[x].h3.a.attrs['title'].replace(',','')
            c=s2[x].p.text
            
            f.write(a+','+b+','+c+'\n')
```
